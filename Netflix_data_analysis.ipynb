```python
df['date_added'].dt.year
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['year_added'].value_counts().sort_index()
```

```python
df['year_added'].value_counts().sort_index().plot(marker='o')
plt.title('Trend of Content Added Over Years')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.show()
```

```python
# it increases gradually from the 2008 to 2019 and from that it slowly reduces little
# and the maximum movies and tv shows are released in the 2018 year
```

```python
plt.figure(figsize=(22,5))
ratings_count = df['rating'].value_counts()
sns.barplot(x=ratings_count.index, y=ratings_count.values, palette='viridis')
plt.title('Most Common Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#most of the ratings are TV-MA then TV-14
```
